```python
numbers = [12, 4, 27, 3, 188, 50, 7]

def find_largest_number(numbers):
    """Returns the largest number from a list of numbers."""
    if not numbers:
        return "no numbers were found"
    largest = numbers[0]
    for number in numbers[1:]:
        if number > largest:
            largest = number
    return largest

print("The largest number is:", find_largest_number(numbers))
```
